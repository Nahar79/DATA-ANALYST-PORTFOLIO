4 - Which country did the most number of Movies and TV Shows?

```python
data.head(2)
```

```python
data['country'].value_counts().head(10)
```

```python
plt.figure(figsize = (12,6))
sns.countplot(y = 'country', order = data['country'].value_counts().index[0:10], data = data)
plt.title('Top 10 Countries with most content in NetFlix')
```

```python
Movie_countries = data[data['type'] == 'Movie']
plt.figure(figsize = (12,6))
sns.countplot(y = 'country', order = data['country'].value_counts().index[0:10], data = Movie_countries)
plt.title('Top 10 Movie Produsing Countries')
```

```python
TV_Shows_Countries = data[data['type'] == 'TV Show']
plt.figure(figsize = (12,6))
sns.countplot(y = 'country', order = data['country'].value_counts().index[0:10], data = TV_Shows_Countries)
plt.title('Top 10 TV Show producing Countries')
```

```python
Director_USA = data[data['country'] == 'United States']
Director_USA['director'].value_counts().head(10)
```

```python
Director_USA = data[data['country'] == 'United States']
Director_USA['director'].value_counts().head(10)
filtered_directors = Director_USA[Director_USA.director != 'Not Given'].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize = (12,6))
sns.countplot(y = filtered_directors, order = filtered_directors.value_counts().index[0:5], data = Director_USA)
plt.title("Top 10 Directors in USA")
plt.show()
```
